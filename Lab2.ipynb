```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

X = np.concatenate((np.random.normal([0,-1],[1,1],[200,2]), np.random.normal([1,1],[1,1],[200,2])))
Y = np.concatenate((np.array([0]*200),np.array([1]*200)))

X1 = np.concatenate((np.random.normal([0,-1],[1,1],[5,2]), np.random.normal([1,1],[1,1],[5,2])))
Y1 = np.concatenate((np.array([0]*5),np.array([1]*5)))

neuron_1 = Perceptron(tol=1e-3, max_iter = 20)
neuron_1.fit(X, Y)
print(neuron_1.score(X1, Y1))
x1 = np.linspace(-4,4,200)
separation_1 = -(1./neuron_1.coef_[0][1])*(neuron_1.coef_[0][0]*x1+neuron_1.intercept_[0])
plt.plot(x1, separation_1, '-r')
plt.scatter(X[:,0], X[:,1], c=Y)
print(neuron_1.coef_)
print(1./neuron_1.coef_[0][1], neuron_1.coef_[0][0], neuron_1.intercept_[0])
```
